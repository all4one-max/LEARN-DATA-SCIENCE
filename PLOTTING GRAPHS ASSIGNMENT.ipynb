### PROBLEM : PLOT JOB_YEAR

```python
import csv,numpy as np,matplotlib.pyplot as plt
with open('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    year=[]
    for row in file_data:
        year.append(row['Posting_date'][-4::])
    year=np.array(year,dtype=int)
    year,posting=np.unique(year,return_counts=True)
    plt.plot(posting,year)
    plt.xlabel('Jobs')
    plt.ylabel('Year')
    plt.title('Amazon Jobs Dataset')
    plt.xticks(rotation=40)
    for i in range(len(year)):
        print(year[i],posting[i])
    plt.show()   
        
```

```python
import csv,pandas as pd,matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8")
posting=df.Posting_date
year=posting.str.split(',',expand=True)[1]
freq=year.value_counts(ascending=True)
freq.sort_index(inplace=True)
for i in range(len(freq)):
    print(int(freq.index[i]),freq.values[i])
```

### PLOT CITY_JOB

```python
import csv,numpy as np,matplotlib.pyplot as plt
with open('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    city=[]
    for row in file_data:
        s=row['location']
        s=[ele for ele in s.strip().split(',')]        
        if s[0]=='IN':
            city.append(s[-1])    
    city=np.array(city)         
    city,count=np.unique(city,return_counts=True)  
    sm=count.sum()
    per=list(map( lambda x:format(((x/sm)*100),'.2f'),count))
    for i in range(len(city)):
        print(city[i],per[i])
```

```python
import csv,pandas as pd,matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8")
lctn=df.location[df.location!='']
lctn=lctn.str.split(',',expand=True)
lctn=lctn[lctn[0]=='IN']
lctn=lctn[2]
opn=lctn.value_counts()
opn.sort_values()
sm=opn.values.sum()
per=list(map( lambda x:format(((x/sm)*100),'.2f'),opn.values))
plt.pie(opn.values,labels=opn.index,autopct='%.2f%%')
plt.show()
for i in range(len(per)):
    print(opn.index[i],str(per[i])+'%')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
file_obj = open('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8")

data = pd.read_csv(file_obj)

locat = data.location
city= locat[locat.str.split(', ',expand = True)[0]=='IN'].str.split(', ',expand = True)[2]
freq = city.value_counts()

plt.pie(freq,labels = freq.index,autopct='%.2f%%')
plt.show()
x = freq.index
perc = np.true_divide(freq, freq.sum(axis=0))*100
for i in range(len(x)):
    print(x[i],format(perc[i],'.2f'))
```

### PLOT JOB_JAVA

```python
import csv,numpy as np,matplotlib.pyplot as plt
with open('C:\\Users\\jhasa\\Desktop\\amazon_jobs_dataset.csv',encoding="utf8") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    year=[]
    for row in file_data:
        if (('Java' in row['BASIC QUALIFICATIONS']) or ('java' in row['BASIC QUALIFICATIONS'])):
            year.append(row['Posting_date'][-4::])
    year=np.array(year)
    year,count=np.unique(year,return_counts=True)
    plt.scatter(year,count,marker='^',linewidth=5)
    plt.show()
    for i in range(len(year)):
        print(int(year[i]),count[i])
```
