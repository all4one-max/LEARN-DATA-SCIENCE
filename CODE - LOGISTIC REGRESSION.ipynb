```python
def step_gradient(data_train,y_data,learning_rate,N,M,m):
    slope_m=[0]*M
    for k in range(M):
        for i in range(N):
            y,fct=y_data[i],0
            for j in range(M):
                fct+=m[j]*data_train.iloc[i,j]            
            val=1/(1+np.exp(-1*fct))
            vl=(-1/N)*data_train.iloc[i,k]*(y-val)
            slope_m[k]+=vl
    for i in range(M):
        m[i]=m[i]-learning_rate*(slope_m[i])
    return m            

def gd(data_train,y_data,learning_rate,no_of_iterations,N,M):   
    m=[0 for i in range(M)]
    for i in range(no_of_iterations):
        m=step_gradient(data_train,y_data,learning_rate,N,M,m)
        cst=cost(data_train,y_data,m,N,M)
        print(i,cst)
    return m

def cost(data_train,y_data,m,N,M):
    cst=0
    for k in range(M):
        for i in range(N):
            y,fct=y_data[i],0
            for j in range(M):
                fct+=m[j]*data_train.iloc[i,j]
            expo=np.exp(fct)
            val=math.log(1+expo)-y*fct
            cst+=val
    return cst

learning_rate=0.2
no_of_iterations=400
N=len(data_train)
M=len(data_train.loc[1,:])
m=gd(data_train,y_data,learning_rate,no_of_iterations,N,M)
```
