```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
oli = datasets.fetch_olivetti_faces()
```

In Olivetti dataset we used are ten different images of each of 40 distinct subjects. For some subjects, the images were taken at different times, varying the lighting, facial expressions (open / closed eyes, smiling / not smiling) and facial details (glasses / no glasses). All the images were taken against a dark homogeneous background with the subjects in an upright, frontal position (with tolerance for some side movement).

```python
fig = plt.figure(figsize=(8, 8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(oli['images'][i], cmap = plt.cm.plasma)
plt.show()
```

```python
x = oli['data']
y = oli['target']
x, y
```

```python
pca = PCA()
pca.fit_transform(x)
```

```python
pca.components_.shape #NOW ORIGINALLY THE DIMENSIONS SHOULD BE 4096 X 4096 BUT WE ARE GETTING 400 X 400
# THE REASON BEING THAT BY DEFAULT THE NUMBER OF COMPONENTS IS THE MINIMUM OF THE DIMENSION OF THE DATASET WE HAVE THAT'S
# WHY MIN(400,4096)=400
```

```python
curr=0
k=0
total=pca.explained_variance_.sum()
while curr/total<0.99:
    curr+=pca.explained_variance_[k]
    k+=1
k
```

```python
pca = PCA(n_components = k)
transfromed_data = pca.fit_transform(x)
transfromed_data.shape
```

```python
x_approx = pca.inverse_transform(transfromed_data)
```

```python
x_approx_images = x_approx.reshape((400,64,64))
```

```python
fig = plt.figure(figsize=(8, 8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(x_approx_images[i], cmap = plt.cm.plasma)
plt.show()
```

```python
eigvec = pca.components_
eigfaces = eigvec.reshape((260,64,64))
```

```python
fig = plt.figure(figsize=(8, 8))
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(eigfaces[i], cmap = plt.cm.plasma)
plt.show()
```
