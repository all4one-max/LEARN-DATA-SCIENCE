### IMDB PROFIT

```python
import sqlite3,pandas as pd
db=sqlite3.connect('IMDB.sqlite')
cur=db.cursor()
```

```python
row=cur.execute('select earning.movie_id,earning.domestic,earning.worldwide,imdb.title from earning join imdb on earning.movie_id=imdb.movie_id where imdb.title!=""')
```

```python
li=[]
for ele in row:
    li.append(ele)    
```

```python
df=pd.DataFrame(li,columns=["movie_id","domestic","worldwide","Title"])
df["total"]=df["domestic"]+df["worldwide"]
df1=df.loc[:,["Title","total"]]
arr=df1.total.values
ind=arr.argsort()
print(df.iloc[ind[-1],:])
```

### IMDB LEAST BUDGET

```python
import sqlite3,pandas as pd
db=sqlite3.connect('IMDB.sqlite')
cur=db.cursor()
```

```python
row=cur.execute('select earning.movie_id,earning.domestic,earning.worldwide,imdb.title,imdb.budget from earning join imdb on earning.movie_id=imdb.movie_id where imdb.title!=""')
```

```python
li=[]
for ele in row:
    li.append(ele)   
```

```python
df=pd.DataFrame(li,columns=["movie_id","domestic","worldwide","Title","budget"])
df['budget']=pd.to_numeric(df["budget"])
df["net_profit"]=df["domestic"]+df["worldwide"]-df["budget"]
df.drop(["domestic","worldwide"],axis=1,inplace=True)
df.sort_values("budget",inplace=True)
df2=df[df.budget==1000000.0]
df2.sort_values("net_profit",inplace=True,ascending=False)
df2
```

```python
db.close()
```

### IMDB YEAR TOP

```python
import sqlite3,pandas as pd
db=sqlite3.connect('IMDB.sqlite')
cur=db.cursor()
```

```python
row=cur.execute('select earning.domestic+earning.worldwide as total,imdb.title from earning join imdb on earning.movie_id=imdb.movie_id where imdb.title!=""')
```

```python
li=[]
for ele in row:
    li.append(ele)  
```

```python
df=pd.DataFrame(li,columns=["total","Title"])
df['year']=df.Title.str.split("(",expand=True)[1]
df['Titles']=df.Title.str.split("(20)",expand=True)[0]
df['year2']=df.Title.str.split("(",expand=True)[2]
df.year2.fillna(df.year,inplace=True)
df.drop('year',axis=1,inplace=True)
df.drop('Title',axis=1,inplace=True)
df['Titles']=df.Titles.str.strip('(')
df['year2']=df.year2.str.strip(')')
df.sort_values("year2",inplace=True)
df['total']=pd.to_numeric(df['total'])
df.reset_index(inplace=True,drop=True)
twenty0=df.Titles[df.total[df['year2']=="2010"].argmax()]
twenty1=df.Titles[df.total[df['year2']=="2011"].argmax()]
twenty2=df.Titles[df.total[df['year2']=="2012"].argmax()]
twenty3=df.Titles[df.total[df['year2']=="2013"].argmax()]
twenty4=df.Titles[df.total[df['year2']=="2014"].argmax()]
twenty5=df.Titles[df.total[df['year2']=="2015"].argmax()]
twenty6=df.Titles[df.total[df['year2']=="2016"].argmax()]
print(2010,twenty0)
print(2011,twenty1)
print(2012,twenty2)
print(2013,twenty3)
print(2014,twenty4)
print(2015,twenty5)
print(2016,twenty6)
```

### IMDB GENRE BUDGET

```python
import sqlite3,pandas as pd,matplotlib.pyplot as plt,numpy as np
db=sqlite3.connect('IMDB.sqlite')
cur=db.cursor()
```

```python
row=cur.execute('select genre.genre,sum(imdb.budget) from genre inner join imdb on genre.movie_id=imdb.movie_id where genre.genre!="" group by genre.genre')
```

```python
li=[]
for ele in row:
    li.append(ele)
```

```python
df=pd.DataFrame(li,columns=["genre","budget"])
df['budget']=pd.to_numeric(df.budget)
df.budget.fillna(0,inplace=True)
label=df.genre.values
val=df.budget.values
plt.pie(val,labels=label,autopct="%.2f%%")
plt.axis("equal")
plt.show()
sm=val.sum()
result=np.true_divide(val*100,sm)
result
```
