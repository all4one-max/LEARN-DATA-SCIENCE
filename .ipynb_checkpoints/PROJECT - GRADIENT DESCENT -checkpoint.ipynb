### QUESTION 1 - BOSTON DATASET

```python
from sklearn import datasets
from sklearn import preprocessing
import pandas as pd,numpy as np
```

```python
data_train=np.genfromtxt("C:\\Users\\jhasa\\Desktop\\gd1_train.csv",delimiter=",")
```

```python
data_test=np.genfromtxt("C:\\Users\\jhasa\\Desktop\\gd1_test.csv",delimiter=",")
```

```python
data_train=data_train[1:,:]
```

```python
y_data=data_train[:,13]
data_train=data_train[:,0:13]
data_train.shape,y_data.shape
```

```python
scaler=preprocessing.StandardScaler()
scaler.fit(data_train)
data_train=scaler.transform(data_train)
```

```python
data_test=scaler.transform(data_test)
```

```python
add=np.ones(378).reshape((378,1))
data_train=np.concatenate((data_train,add),axis=1)
data_train.shape
```

```python
add=np.ones(127).reshape((127,1))
data_test=np.concatenate((data_test,add),axis=1)
data_test.shape
```

```python
def step_gradient(points,learning_rate,m,y_data):
    N=len(points)
    M=len(points[0])
    slope_m=[0]*M  
    for k in range(M):
        for i in range(N):
            fct,y=0,y_data[i]
            for j in range(M):
                fct=fct+m[j]*points[i,j]
            slope_m[k]+=(-2/N)*(y-fct)*points[i,k]
    for i in range(M):
        m[i]=m[i]-learning_rate*slope_m[i]
    return m

def gd(points,learning_rate,no_iterations,y_data):
    N=len(points)
    m=[0]*(len(points[0]))
    for i in range(no_iterations):
        m=step_gradient(points,learning_rate,m,y_data)
        cst=cost(points,m,y_data)
        print(i,cst)
    return m

def cost(points,m,y_data):
    N=len(points)
    M=len(points[0])
    cst=0
    for i in range(N):
        y=y_data[i]
        fct=0
        for j in range(M):
            fct=fct+m[j]*points[i,j]
        cst+=(1/N*((y-fct)**2))
    return cst

learning_rate=0.16
no_iterations=500
m=gd(data_train,learning_rate,no_iterations,y_data)
```

```python
m=np.array(m)
m.shape

y_test=[0]*127

N,M=127,14
for i in range(N):
    y=0
    for j in range(M):
        y+=data_test[i,j]*m[j]
    y_test[i]=y
    
y_test=np.array(y_test)
```

```python
np.savetxt(r'C:\\Users\\jhasa\\Desktop\\gd1_prediction.csv', y_test,fmt='%.15f')
```

```python
from sklearn import datasets
from sklearn import preprocessing
import pandas as pd,numpy as np

def step_gradient(points,learning_rate,m,y_data):
    N=len(points)
    M=len(points[0])
    slope_m=[0]*M  
    for k in range(M):
        for i in range(N):
            fct,y=0,y_data[i]
            for j in range(M):
                fct=fct+m[j]*points[i,j]
            slope_m[k]+=(-2/N)*(y-fct)*points[i,k]
    for i in range(M):
        m[i]=m[i]-learning_rate*slope_m[i]
    return m

def gd(points,learning_rate,no_iterations,y_data):
    N=len(points)
    m=[0]*(len(points[0]))
    for i in range(no_iterations):
        m=step_gradient(points,learning_rate,m,y_data)
        cst=cost(points,m,y_data)
        print(i,cst)
    return m

def cost(points,m,y_data):
    N=len(points)
    M=len(points[0])
    cst=0
    for i in range(N):
        y=y_data[i]
        fct=0
        for j in range(M):
            fct=fct+m[j]*points[i,j]
        cst+=(1/N*((y-fct)**2))
    return cst

data_train=np.genfromtxt("C:\\Users\\jhasa\\Desktop\\gd2_train.csv",delimiter=",")

data_test=np.genfromtxt("C:\\Users\\jhasa\\Desktop\\gd2_test.csv",delimiter=",")

data_train=data_train[1:,:]
y_data=data_train[:,4]
data_train=data_train[:,0:4]

scaler=preprocessing.StandardScaler()
scaler.fit(data_train)
data_train=scaler.transform(data_train)

data_test=scaler.transform(data_test)

add=np.ones(7175).reshape((7175,1))
data_train=np.concatenate((data_train,add),axis=1)

add=np.ones(2392).reshape((2392,1))
data_test=np.concatenate((data_test,add),axis=1)

learning_rate=0.1
no_iterations=115

m=gd(data_train,learning_rate,no_iterations,y_data)
print(m)
```

```python
m=np.array(m)
m.shape

y;_test=[0]*2392

N,M=2392,5
for i in range(N):
    y=0
    for j in range(M):
        y+=data_test[i,j]*m[j]
    y_test[i]=y
    
y_test=np.array(y_test)
```
