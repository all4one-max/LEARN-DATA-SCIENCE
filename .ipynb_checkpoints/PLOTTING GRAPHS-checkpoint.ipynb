### CUSTOMISING GRAPH

```python
import numpy as sj
import matplotlib.pyplot as plt
x=sj.arange(1,5,0.1)
y=x**3
z=x**4
plt.plot(x,y,color='red',marker='o')
plt.plot(x,z,color='green',marker='p')
plt.show()
```

```python
import numpy as sj
import matplotlib.pyplot as plt
x=sj.arange(1,5,0.1)
y=x**3
z=x**2
plt.ylabel('y')
plt.xlabel('x')
plt.title('MatPlotlib Demo')
plt.plot(x,y,'r',linewidth=5,label='x^3')
plt.plot(x,z,'g',linewidth=5,label='x^2')
plt.axis([0,10,0,100])
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as sj
import matplotlib.pyplot as plt
x=sj.arange(1,5,0.1)
y=x**3
z=x**2
plt.ylabel('y')
plt.xlabel('x')
plt.title('MatPlotlib Demo')
plt.plot(x,y,c='r',linewidth=5,label='x^3')
plt.plot(x,z,'g',linewidth=5,label='x^2')
plt.text(2,80,'(2,80)',fontsize=13)
plt.grid()
plt.legend()
plt.show()
```

### BUBBLE GRAPH

```python
import matplotlib.pyplot as plt
year=[2012,2013,2014,2015,2016,2017]
avg_salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.scatter(year,avg_salary,color='yellow',s=population,edgecolor='r')
plt.show()
```

```python
plt.scatter(year,avg_salary,c='yellow',s=population,edgecolor='r',alpha=0.5)               
plt.show() # alpha determines how opaque are point will be
```

```python
import numpy as np
pop_npy=np.array(population)
plt.scatter(year,avg_salary,s=population,c=pop_npy,marker='^')    
plt.show()
```

```python
import numpy as np
pop_npy=np.array(population)
plt.scatter(year,avg_salary,s=population,c=pop_npy,marker='^')  
plt.text(year[1]+0.05,avg_salary[1]+0.05,population[1])
plt.show()
```

### MICROSOFT BUBBLE GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np
employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
emp=np.arange(len(year))
plt.scatter(year,revenue,s=employees,marker='^',c=emp)
plt.show()
print(2008,60.42,91)
print(2015,93.58,118)
```

### PIE CHART

```python
import matplotlib.pyplot as plt
sizes=[3,4,6,2]
color=['r','g','blue','yellow']
plt.pie(sizes,colors=color)
plt.show()
```

```python
label=['A','B','C','D']
plt.title('PIE')
exploded=[0.1,0.2,0.4,0.3]
plt.pie(sizes,labels=label,explode=exploded)
plt.show()
```

```python
label=['A','B','C','D']
plt.pie(sizes,labels=label,autopct='%0.2f%%',counterclock=False,startangle=100)
plt.axis('equal')
plt.show()
```

```python
label=['A','B','C','D']
plt.pie(sizes,labels=label,autopct='%0.2f%%',startangle=100)
plt.show()
```

### LAPTOP SALES

```python
import matplotlib.pyplot as plt
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
plt.pie(sold,labels=company,autopct='%0.1f')
sm=sum(sold)
plt.show()
x=list(map (lambda x:format((x/sm)*100,'.1f'),sold))
for i in range(len(x)):
    print(company[i],x[i])
```

```python
import matplotlib.pyplot as plt
import numpy as np
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
company=np.array(company)
sold=np.array(sold)
plt.pie(sold,labels=company,autopct='%0.1f')
x=np.true_divide(sold,sold.sum())*100
plt.show()
for i in range(len(x)):
    print(company[i],format(x[i],'0.1f'))
```
