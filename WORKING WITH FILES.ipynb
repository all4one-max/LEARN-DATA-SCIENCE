### READING CSV FILE USING CSV MODULE

```python
import csv # csv.reader allows us to read the content of the file line by line
with open('C:\\Users\\jhasa\\Desktop\\csv.csv') as file_obj:
    file_data=csv.reader(file_obj)
    print(type(file_data)) # earlier file_data was a list type object now it's under the class csv.reader
    print()    
    for row in file_data: # you cannot slice file_data as it is not a list hence we can use counter to get the required
        print(row)        # number of rows
        
```

### PROBLEM: PRINT FIRST THREE LINES

```python
import csv
with open('C:\\Users\\jhasa\\Desktop\\year2017.csv') as file_obj:
    file_data=csv.reader(file_obj)
    rw=1
    for row in file_data:
        if rw!=1:
            for ele in row:
                print(ele,end=" ")
            print()    
        rw+=1
        if rw==5:
            break
```

### PROBLEM : TOTAL WOUNDED PEOPLE

```python
import csv
with open('C:\\Users\\jhasa\\Desktop\\year2017.csv') as file_obj:
    file_data=csv.reader(file_obj)
    data=list(file_data)
summ=0
for i in range(1,len(data)):
    val=data[i][10]
    if val!='':
        summ+=float(val)
print(int(summ))     
```

### PROBLEM : TOTAL WOUNDED PEOPLE FROM INDIA

```python
import csv
with open('C:\\Users\\jhasa\\Desktop\\year2017.csv') as file_obj:
    file_data=csv.reader(file_obj)
    data=list(file_data)
summ=0
for i in range(1,len(data)):
    val=data[i][10]
    if val!='' and data[i][3]=='India':
        summ+=float(val)
print(int(summ))  
```
