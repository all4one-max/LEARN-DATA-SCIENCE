```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### VISUALIZING STATISTICAL RELATIONSHIP

```python
sns.set()
df = sns.load_dataset('tips')
df.head()
```

```python
plt.scatter(df.total_bill,df.tip)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,style='time',hue='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='time',col='smoker')
plt.show()
```

### PROBLEM - MPG DATASET

```python
mpg=sns.load_dataset('mpg')
mpg.head()
```

```python
sns.relplot(x='mpg',y='horsepower',data=mpg)
plt.show()
```

### LINEPLOT

```python
flight = sns.load_dataset('flights')
flight.head()
```

```python
plt.plot(flight.year,flight.passengers)
plt.show()
```

```python
sns.relplot(kind='line',data=flight,x='year',y='passengers')
plt.show()
```

```python
sns.relplot(kind='line',data=flight,x='year',y='passengers',hue='month')
plt.show()
```

### VISUALIZING CATEGORICAL DATA

```python
tip = sns.load_dataset('tips')
tip.head()
```

```python
sns.catplot(x='sex',y='total_bill',data=tip) # default value of jitter is 1
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tip,jitter=0.25)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tip,kind='swarm')
plt.show()
```

### CATEGORICAL DISTRIBUTION PLOT

```python
sns.catplot(x='sex',y='total_bill',data=tip,kind='box',hue='smoker')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tip,kind='boxen',hue='smoker')
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.catplot(x='species',y='sepal_width',data=iris,kind='violin')
plt.show()
```
