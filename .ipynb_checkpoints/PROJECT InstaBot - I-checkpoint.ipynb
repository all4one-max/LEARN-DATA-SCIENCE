```python
# fetching how many people follow me and how many i follow

follower_following=driver.find_elements_by_xpath("//a[@class='-nal3 ']/span")
no_of_followers=int(follower_following[0].text)
no_of_following=int(follower_following[1].text)
```

```python
# HERE I FOUND THE INSTANT OF FOLLOWERS BOX THAT OPENED to fetch the insta handles of my followers
driver.find_element_by_partial_link_text("followers").click()

follower_box=driver.find_element_by_class_name("isgrP")

scroll=0 # "scroll" IS THE NUMBER OF DIFFERENT NAMES FOUND
while scroll<no_of_followers: 
    scroll = len(follower_box.find_elements_by_class_name("d7ByH"))
    my_follower_list=follower_box.find_elements_by_class_name("d7ByH") 
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",follower_box)    
   
follower_list=[]
for ele in my_follower_list:
    follower_list.append(ele.text)
```

```python
# CLOSING THE FOLLOWERS BOX BY CLICKING ON "X" BUTTON

driver.find_element_by_xpath("//div[@class='WaOAr']/button").click()
```

```python
# HERE I FOUND THE INSTANT OF FOLLOWERS BOX THAT OPENED to fetch the insta handles of people that follow me
driver.find_element_by_partial_link_text("following").click()

following_box=driver.find_element_by_class_name("isgrP")

scroll=0 # "scroll" IS THE NUMBER OF DIFFERENT NAMES FOUND
while scroll<no_of_following: 
    scroll = len(following_box.find_elements_by_class_name("d7ByH"))
    my_following_list=following_box.find_elements_by_class_name("d7ByH") 
    time.sleep(1)
    driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",following_box)    
    
my_following=[]
for ele in my_following_list:
    my_following.append(ele.text)
```

```python
# here i wil print those followers of "foodtalkindia" who i follow but they don't follow me back

found=False
for ele in foodtakindia_followers:
    if ele in my_following and ele not in my_following:
        found=True
        print(ele)
if not(found):
    print("THERE ISN'T A SINGLE INSTA HANDLE THAT SATISFY THE CONDITION GIVEN IN THE QUESTION")
```
