### PROBLEM : IRIS FEATURE

```python
import pandas as pd
column_name=['sl','sw','pl','pw','flower_type']
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=column_name)
df=iris.copy()
df1=df[df.flower_type=='Iris-virginica']
df2=df[df.flower_type=='Iris-versicolor']
df3=df[df.flower_type=='Iris-setosa']
a1=df3.describe().values
a2=df2.describe().values
a3=df1.describe().values
a= list(map(lambda x: format(x,'.2f'), a1[3]))
print(*a,'Iris-setosa')
a= list(map(lambda x: format(x,'.2f'), a1[7]))
print(*a,'Iris-setosa')
a= list(map(lambda x: format(x,'.2f'), a1[1]))
print(*a,'Iris-setosa')
a= list(map(lambda x: format(x,'.2f'), a2[3]))
print(*a,'Iris-versicolor')
a= list(map(lambda x: format(x,'.2f'), a2[7]))
print(*a,'Iris-versicolor')
a= list(map(lambda x: format(x,'.2f'), a2[1]))
print(*a,'Iris-versicolor')
a= list(map(lambda x: format(x,'.2f'), a3[3]))
print(*a,'Iris-virginica')
a= list(map(lambda x: format(x,'.2f'), a3[7]))
print(*a,'Iris-virginica')
a= list(map(lambda x: format(x,'.2f'), a3[1]))
print(*a,'Iris-virginica')
```

### HANDLING NAN

```python
import pandas as pd
import numpy as sj
column_name=['sl','sw','pl','pw','flower_type']
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=column_name)
df=iris.copy()
df.iloc[2:4,1:3]=sj.nan
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.iloc[2:4,1:3]=sj.nan
df.head()
```

```python
df.sw.fillna(df.sw.mean(),inplace=True)
df.head()
```

```python
df1=df[df.flower_type=='Iris-setosa']
x=df1.pl.mean()
df.pl.fillna(x,inplace=True)
df.head()
```

### HANDLING STRINGS IN DATA

```python
import pandas as pd
import numpy as sj
column_name=['sl','sw','pl','pw','flower_type']
iris=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',names=column_name)
df=iris.copy()
df['Gender']='Female'
df.iloc[3:10,5]='Male'
df.head(10)
```

```python
def f(s):
    if s=='Male':
        return 0
    return 1

df['sex']=df.Gender.apply(f)
df.head(10)
```

```python
df.drop('Gender',axis=1,inplace=True)
df.head(10)
```
